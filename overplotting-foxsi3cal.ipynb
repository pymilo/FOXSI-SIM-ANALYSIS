# Simulation vs. Data
### February 2017

```python
%matplotlib inline
import matplotlib.pyplot as plt
from astropy.io import fits
```

# 26 arcmin

```python
#DATA
hdulist = fits.open('FOXSI3Data/Foxsi_mod6_26arcmin_0.5mA.fits')
hdu = hdulist[0]
#SIMULATION
simu = fits.open('26.0arcmin100000.fits')
sim = simu[0]
plt.title('26 arcmin - Simulation')
plt.imshow(sim.data[320:710,380:770], origin='lower',cmap='Reds',vmin=0.5)
```

```python
plt.title('26 arcmin - Data')
plt.imshow(hdu.data[0,:,:], origin='lower',cmap='Blues',vmin=1350,vmax=1400)
```

```python
fig, ax = plt.subplots()
ax.set_title('26 arcmin - Overplot')
ax.imshow(sim.data[320:710,380:770], origin='lower',cmap='Reds',vmin=0.5,
          extent=(0,1024,0,1024),alpha=0.5)
ax.imshow(hdu.data[0,:,:], origin='lower',cmap='Blues',vmin=1300,vmax=1350,alpha=0.5)
```
